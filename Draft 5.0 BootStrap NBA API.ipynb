<h3>This is an example of how to generate lineups for a NBA Showdown contest for DraftKings.<h3>

```python
from fantasy import *
```

Import the files for a given game.

```python
draft, players = draftMerge('Wizards','Grizzlies', 'Mar 10')
```

```python
draft
```

```python
import ipywidgets as widgets

def playerButtonArray(players_array):

    switch = [widgets.ToggleButton(description = name, value = False) for name in players_array]
    row1 = widgets.VBox([items for items in switch[:8]])
    row2 = widgets.VBox([items for items in switch[8:16]])
    row3 = widgets.VBox([items for items in switch[16:24]])
    row4 = widgets.VBox([items for items in switch[24:32]])
    row5 = widgets.VBox([items for items in switch[32:]])
    combined = widgets.HBox([row1, row2, row3, row4, row5])
    display(combined)
    
    def upon_clicked(btn):
        global pool

        signal = btn.owner.value
        desc = btn.owner.description
        if signal == True:
            pool.append(desc)
        else:
            try:
                pool = remAll(pool, desc)
            except:
                pass
    
    for n in range(len(players_array)):
        switch[n].observe(upon_clicked, names = ['value'])
    
    return pool
hide_toggle()
```

Choose players that you want to create lineups with.

```python
pool = []
pool = playerButtonArray(players)
```

Generate all possible lineups with the given selected pool.

```python
lineups = generateLineups1(pool, draft)
lineups.describe()
```

```python
def coreButtonArray(pool):
    switch = [widgets.ToggleButton(description = name, value = False) for name in pool]
    row1 = widgets.VBox([items for items in switch[:5]])
    row2 = widgets.VBox([items for items in switch[5:10]])
    row3 = widgets.VBox([items for items in switch[10:15]])
    row4 = widgets.VBox([items for items in switch[15:]])
    combined = widgets.HBox([row1, row2, row3, row4])
    display(combined)
    
    def upon_clicked(btn):
        global core
        signal = btn.owner.value
        desc = btn.owner.description
        if signal == True:
            core.append(desc)
        else:
            try:
                core = remAll(core, desc)
            except:
                pass
    
    for n in range(len(pool)):
        switch[n].observe(upon_clicked, names = ['value'])
    
    return core

hide_toggle()
```

Choose players that you want the lineups to be largely dominated with.

```python
core = []
core = coreButtonArray(pool)
```

```python
lineups_filtered = filterCorePlayers_noteam(core, lineups)
```

Simulate the game matchup 1,000 times through random sampling. 
The arbitrarily determined weight are as follows: 
- 30% from data this season
- 20% from data of last 10 games
- 20% from data last season
- 10% data from last last season
- 10% data from this season dependent on home/away venue
- 10% data from history with opponent

```python
sim_games, min_dict_10, min_dict_5 = simulateGames('Grizzlies', 'MEM', 'Wizards', 'WAS', lineups_filtered, draft, pool)
```

Three Functions:

- Attach the probabilities that each lineup reaches both 200 points and 250 points
- Attach the average minutes that each lineup is on the floor, minutes come from game simulation in cell prior
- Attach the combined (modified) usage rate of each lineup

```python
lineups_filtered = valueProbability(lineups_filtered, sim_games)
lineups_filtered = minPlayed(lineups_filtered, min_dict_5, min_dict_10)
lineups_filtered = usageMetric(lineups_filtered, players)
```

Sort lineups by probability of reaching 200 points, then by combined usage rate of lineups. This is just one of many potential ways to sort a lineup.

```python
lineups_filtered.sort_values('Probs 4x', ascending = False).head(100).sort_values('L10 Metric', ascending = False).head(10)
```

Export the Lineups to a Local Drive

```python
lineups.to_csv('Games/Mar 10/WASatMEM_Mar 10_D.csv')
lineups_filtered.to_csv('Games/Mar 10/WASatMEM_Mar 10_DF.csv')
```
